```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter
```

```python
df_kpbulanan = pd.read_csv("/content/kebutuhan_per_bulan_2.csv")
```

```python
df_kpbulanan
```

Scater plot 1

```python
df = pd.DataFrame(df_kpbulanan)

provinsi_tertentu = 'ACEH'
df_provinsi_tertentu = df.loc[df['Propinsi'] == provinsi_tertentu]

x = df_provinsi_tertentu['tahun']
y = df_provinsi_tertentu['kebutuhan']

plt.title(f'STATISTIK KEBUTUHAN {provinsi_tertentu}')
plt.xlabel("Tahun")
plt.ylabel("Kebutuhan")

plt.scatter(x, y, color='#D04', marker='s')

plt.xlim(min(x), max(x))


plt.show()
```
